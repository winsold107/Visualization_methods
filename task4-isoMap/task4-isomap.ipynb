В качестве бонуса можно реализовать Hessian LLE или UMAP. 

 

```python
from sklearn import datasets
from sklearn.neighbors import NearestNeighbors

from matplotlib import pyplot as plt

import numpy as np

import networkx as nx

from torch import randn, Tensor
from torch.nn.functional import pdist
from torch.nn.parameter import Parameter
from torch.nn.init import normal_
from torch.linalg import norm
from torch.optim import SGD
from numpy import linalg
import scipy
```

```python
class IsoMap:
    def __init__(self, n = 1000, k = 5, m = 2):
        self.n = n
        self.k = k+1
        self.X, self.colors = datasets.make_s_curve(n_samples=n)
        self.G = nx.Graph()
        self.m = m
        
        
    def get_neighbours_arrs(self):
        neighbours = NearestNeighbors(n_neighbors=self.k).fit(self.X)
        neigh_dist, neigh_ind = neighbours.kneighbors(self.X)
        return neigh_dist, neigh_ind

    
    def get_distance_graph(self, neigh_dist, neigh_ind):
        for i in range(self.n):
            for j in range(1, self.k):
                self.G.add_edge(i, neigh_ind[i][j], weight=neigh_dist[i][j])

                
    def get_shortest_dists(self):
        shortest_dists = dict(nx.all_pairs_dijkstra_path_length(self.G))
        return shortest_dists
    
    
    def compute_lowerdim_embed(self, shortest_dists):
        D = np.zeros((self.n, self.n))
        for i in range(self.n-1):
            for j in range(i+1, self.n):
                D[i][j] = (shortest_dists[i][j])**2
                D[j][i] = (shortest_dists[i][j])**2
        
        C = np.diag(np.full(self.n, 1)) - (1/self.n)*np.ones(self.n)
        centered = np.dot(np.dot((-1/2)*C, D), C)
        eig_vals, eig_vecs = linalg.eig(centered)
        ids = np.argsort(eig_vals)[::-1]
        E_vec = eig_vecs[:, ids[0]:ids[0]+1]
        E_val = [eig_vals[ids[0]]]
        for i in range(1, self.m):
            E_vec = np.hstack((E_vec, eig_vecs[:, ids[i]:ids[i]+1]))
            E_val.append(eig_vals[ids[i]])
        return np.dot(E_vec, scipy.linalg.sqrtm(np.diag(E_val)))

    
    def isomap_algo(self):
        dists, inds = self.get_neighbours_arrs()
        self.get_distance_graph(dists, inds)
        shortest_dists = self.get_shortest_dists()
        self.res_matrix = self.compute_lowerdim_embed(shortest_dists)

        
    def plot(self):
        fig = plt.figure(figsize=(10, 10))

        ax1 = fig.add_subplot(211, projection='3d')
        ax1.scatter(self.X[:, 0], self.X[:, 1], self.X[:, 2], c=self.colors, cmap=plt.cm.Spectral)
        ax1.set_title("Multidim data")
        
        ax2 = fig.add_subplot(212)
        ax2.scatter(self.res_matrix[:, 0], self.res_matrix[:, 1], c=self.colors, cmap=plt.cm.Spectral)
        plt.title('Lowered dim data')
        plt.show()
        fig.savefig('result.svg')
```

```python
isomap = IsoMap()
isomap.isomap_algo()
isomap.plot()
```
